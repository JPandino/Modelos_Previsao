Dados com sazonalidade


## Importação das Bibliotecas

```python
# Manipulação dos Dados
import numpy as np
import pandas as pd
from datetime import datetime
# Modelagem
from statsmodels.tsa.api import ExponentialSmoothing
# Visualização de Dados
import matplotlib.pyplot as plt
```

## Importação dos Dados

```python
# Carregamento do dataset
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('/content/drive/MyDrive/Github/AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
data.index.freq = 'MS'
```

## ExponentialSmoothing

```python
fit1 = ExponentialSmoothing(data, seasonal_periods=12, trend='additive', seasonal='additive', use_boxcox=True).fit()
fit1.fittedvalues.plot(style='--', color='red')
plt.show()
```

```python
fit1.forecast(24).plot(style='--', marker='o', color='black', legend=True)
plt.show()
```
